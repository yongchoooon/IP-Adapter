```python
import os
import matplotlib.pyplot as plt
from glob import glob

original_image_path = 'VOCdevkit/VOC2012/JPEGImages'
generated_image_path = 'generated'
model_name = 'plus_sdxl'

genearated_dir_path = os.path.join(generated_image_path, model_name)
```

```python
original_images = sorted([original_image_path + '/' + path for path in os.listdir(original_image_path)])
generated_images = sorted([genearated_dir_path + '/' + path for path in os.listdir(genearated_dir_path)])

print('Original Images 수 :', len(original_images))
```

```python
print('Generated Images 수 :', len(generated_images))

image_index = 16

image_name = '_'.join(generated_images[image_index * 2].split('/')[-1].split('_')[:-1])
print('image_name :', image_name)

original_image_file = glob(original_image_path + '/' + image_name + '*')[0]
generated_image_files = glob(genearated_dir_path + '/' + image_name + '*')

fig, ax = plt.subplots(1, 3, figsize=(21, 7))
ax[0].imshow(plt.imread(original_image_file))
ax[0].set_title('Original Image')
ax[1].imshow(plt.imread(generated_image_files[0]))
ax[1].set_title('Generated Image 1 : seed 42')
ax[2].imshow(plt.imread(generated_image_files[1]))
ax[2].set_title('Generated Image 2 : seed 421')
```
